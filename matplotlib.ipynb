```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
y=x*2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth='10')
```

```python
plt.plot(x,y)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
z=x*3
```

```python
plt.plot(x,y,color='r',linestyle='-',linewidth='10')
plt.plot(x,z,color='g',linestyle=':',linewidth='10')
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='r',linestyle='-',linewidth='10')
plt.subplot(1,2,2)
plt.plot(x,z,color='g',linestyle=':',linewidth='10')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,color='r',linestyle='-',linewidth='10')
plt.subplot(2,1,2)
plt.plot(x,z,color='g',linestyle=':',linewidth='10')
plt.show()
```

# bar plot

```python
stu={'sumit':10,'Ajay':5,'manohar':3,'chandan':8}
name=list(stu.keys())
num=list(stu.values())
```

```python
stu
```

```python
name
```

```python
num
```

```python
plt.bar(name,num)
plt.show()
```

```python
plt.barh(name,num,color='g')
plt.show()
```

# scatter plot

```python
a=[10,20,30,40,50]
b=[3,6,9,12,15]
plt.scatter(a,b)
plt.show()
```

```python
plt.scatter(a,b,marker='*',c='g',s=100)
plt.show()
```

```python
a=[10,20,30,40,50]
b=[3,6,9,12,15]
c=[4,8,12,16,20]
plt.subplot(1,2,1)
plt.scatter(a,b,marker='.',c='r',s=100)
plt.subplot(1,2,2)
plt.scatter(a,c,marker='*',c='g',s=100)
plt.show()
```

# histogram

```python
data=[1,3,3,3,4,4,5,6,7,9,8,9,5]
plt.hist(data)
plt.show()
```

```python
plt.hist(data,color='y',bins=30)
plt.show()
```

# boxplot

```python
one=[10,20,30,40,50]
two=[3,6,9,12,15]
three=[4,8,12,16,20]
dat=[one,two,three]
plt.boxplot(dat)
plt.show()
```

```python
plt.violinplot(dat)
plt.show() 
```


# piechart

```python
fruit=['apple','orange','mango','guava']
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([100],colors=['white'],radius=1)
plt.show()
```
